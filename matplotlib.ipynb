```python
import matplotlib.pyplot as plt




salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]

fig, ax = plt.subplots()

names = list(map(lambda tup:tup[0], salaries))
salary_values = list(map(lambda tup:tup[1], salaries))

ax.barh(names,salary_values,color='g')
```

```python
import plotly.graph_objects as go

salaries = [
	("Mark", 1000),
	("John", 1500),
	("Daniel", 2300),
	("Greg", 5000)
]

group_names = list(map(lambda tup:tup[0], salaries))
group_salary = list(map(lambda tup:tup[1], salaries))


data = go.Bar(x=group_names,y=group_salary)

layout = {
	'title': 'Salaries with plotly'
}
fig = go.Figure(data, layout)
fig.show()
```
